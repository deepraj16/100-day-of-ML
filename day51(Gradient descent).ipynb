```python
import pandas as pd
```

```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
#OLS
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x,y)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
# y(predict) = x(input)* lr.coef_(m)  +lr.intercept(b)
```

```python
#using Gradient desent
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='Red') # the given line for regressin
```

# lets apply Gradient desicts asuming m is constant 78.35
# and assume the starting  value is intercept b=0

```python
y_pred=((78.35*x)+0).reshape(4)
```

```python
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,lr.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='#00a65a',label='b = 0')
plt.legend()
plt.show()
```

```python
m=78.35
b=0
l_rate=0.1 # learing rate
loss_slop = -2 * np.sum(y- m * x.ravel() -b)
loss_slop
```

```python
step_size = loss_slop*l_rate
step_size
```

```python
b = b - step_size
b
```

```python
y_pred1=((78.35*x) + b ).reshape(4)
```

```python
plt. scatter(x,y)
plt.plot(x,lr.predict(x),color='red',label='OLS')
plt.plot(x,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(x,y_pred,color='#A3E4D7',label='b=0')
plt.legend()
plt.show()
```
